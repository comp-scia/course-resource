```python
y = np.reshape(y,(y.shape[0],1))
```

```python
# Train test split 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=.20,random_state=32)
```

1. pick random W and  b
2. make predictions y`
3. calculate J(W,b)
4. use gradient decent to make update
5. repeat until J settles 

```python
# Step 1
W = np.random.randn(2,1) 
b = np.random.randn(1)
```

```python
# Step 2
z = np.dot(x_train, W) + b
# Sigmoid fucntion
# np.exp(n) -> e^(n)
y_predicted = 1.0/(1.0 + np.exp(-1.0 * z))
```

```python
# Step 3
cost = -1 * np.mean((y_train * np.log(y_predicted)) + (1-y_train) * np.log(1-y_predicted))
cost
```

```python
# Step 4
dcost  = (y_predicted - y_train)
dw = (np.dot(x_train.T,dcost))/x_train.shape[0]
db = np.mean(dcost)
learning_rate = 0.01
W = W - learning_rate * dw
b = b - learning_rate * db
```

```python
# Step 5
for i in range(1000):
    z = np.dot(x_train, W) + b
    y_predicted = 1.0/(1.0 + np.exp(-1.0 * z))
    cost = -1 * np.mean((y_train * np.log(y_predicted)) + ((1-y_train) * np.log(1-y_predicted)))
    if i%1==0:
        
        print(i, cost) # print once in every 50 epoch
    dcost  = (y_predicted- y_train)
    dw = (np.dot(x_train.T,dcost))/x_train.shape[0]
    db = np.mean(dcost)
    learning_rate = 0.01
    W = W - learning_rate * dw
    b = b - learning_rate * db
```

```python
def predict(x):
    z = np.dot(x, W) + b
    A = 1.0/(1.0 + np.exp(-1.0 * z))
    
    return A>0.5
```

```python
train_predictions = predict(x_train)
train_accuracy = np.mean(train_predictions==y_train)
```

```python
test_predictions = predict(x_test)
test_accuracy = np.mean(test_predictions==y_test)
```

```python
print(train_accuracy, test_accuracy)
```
